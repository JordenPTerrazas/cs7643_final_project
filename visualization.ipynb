```python
import pandas as pd
import matplotlib.pyplot as plt
from tbparse import SummaryReader
```

```python
# Looking at the baseline
log_dir = "models/baseline/logs/events.out.tfevents.1713876512.DESKTOP-OTEC5N2.5902.0"
reader = SummaryReader(log_dir)
baseline_df = reader.scalars
```

```python
baseline_df.head()
```

```python
loss_train = baseline_df[baseline_df["tag"] == "Loss/train"]
loss_val = baseline_df[baseline_df["tag"] == "Loss/val"]

pesq_train = baseline_df[baseline_df["tag"] == "PESQ/train"]
pesq_val = baseline_df[baseline_df["tag"] == "PESQ/val"]

stoi_train = baseline_df[baseline_df["tag"] == "STOI/train"]
stoi_val = baseline_df[baseline_df["tag"] == "STOI/val"]
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 6))

# Plot validation loss
ax1.plot(loss_val['step'], loss_val['value'])
ax1.set_xlabel('Step')
ax1.set_ylabel('Loss')
ax1.set_title('Validation Loss')

# Plot validation PESQ
ax2.plot(pesq_val['step'], pesq_val['value'])
ax2.set_xlabel('Step')
ax2.set_ylabel('PESQ')
ax2.set_title('Validation PESQ')

plt.tight_layout()
plt.show()
```

```python
fig_stoi, ax_stoi = plt.subplots(figsize=(6, 4))

# Plot validation STOI
ax_stoi.plot(stoi_val['step'], stoi_val['value'])
ax_stoi.set_xlabel('Step')
ax_stoi.set_ylabel('STOI')
ax_stoi.set_title('Validation STOI')

plt.tight_layout()
plt.show()
```
